### XGBoost for multiclass clasification

```python
from sklearn.datasets import make_multilabel_classification
import numpy as np

import xgboost as xgb

# Make a synthetic ranking dataset for demonstration
seed = 1994
X, y = make_multilabel_classification(n_samples=10000, n_features=64, n_classes=100, n_labels=100, length=64, random_state=seed)
```

```python
clf = xgb.XGBClassifier(tree_method="hist", multi_strategy="multi_output_tree")
clf.fit(X, y)
```

```python
np.testing.assert_allclose(clf.predict(X), y, rtol=0, atol=1)
```

```python
scores = clf.predict(X)
scores
```

### XGboost for ranking

```python
from sklearn.datasets import make_classification
import numpy as np

import xgboost as xgb
```

```python
# Make a synthetic ranking dataset for demonstration
seed = 1994
X, y = make_classification(random_state=seed, n_samples=13000, n_classes=1000, n_features=128, n_clusters_per_class=1, n_informative=10)
rng = np.random.default_rng(seed)
n_query_groups = 130
qid = rng.integers(0, n_query_groups, size=X.shape[0])
```

```python
# Sort the inputs based on query index
sorted_idx = np.argsort(qid)
X = X[sorted_idx, :]
y = y[sorted_idx]
qid = qid[sorted_idx]
```

```python
ranker = xgb.XGBRanker(tree_method="hist", lambdarank_num_pair_per_sample=100, objective="rank:ndcg", lambdarank_pair_method="topk", ndcg_exp_gain=False, n_jobs=32)
ranker.fit(X, y, qid=qid)
```

```python
scores = ranker.predict(X)
sorted_idx = np.argsort(scores)[::-1]
scores = scores[sorted_idx]
```
